```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x, y = make_regression(
    n_samples=4,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=80,
    random_state=13
)
```

```python
plt.scatter(x, y)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x, y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(x, y)
plt.plot(x, lr.predict(x), color='red')
```

## Gradient descent

#### Assumption :

##### let slope = 78.35

##### start c = 0


```python
y_pred = ((78.35 * x) + 0).reshape(4)
y_pred
```

```python
plt.scatter(x, y)
plt.plot(x, lr.predict(x), color='r', label='OLS')
plt.plot(x, y_pred, color='y', label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
c = 0


loss_slope = -2 * np.sum(
    y - m*x.ravel() - c
)
loss_slope
```

```python
# learning rate = 0.1

alpha = 0.1
step_size = loss_slope * alpha
step_size
```

```python
c
```

```python
c = c-step_size
```

```python
c
```

```python
y_pred1 = ((78.35 * x) + c).reshape(4)
plt.scatter(x, y)
plt.plot(x, lr.predict(x), color='r', label='OLS')
plt.plot(x, y_pred1, color='y', label='b = 0')
plt.legend()
plt.show()
```

```python
c = -100
m = 78.35
alpha = 0.01

epochs = 100

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*x.ravel() - c)
    c = c - (alpha * loss_slope)
    y_pred = m * x + c
    plt.plot(x, y_pred)

plt.scatter(x, y)
```
